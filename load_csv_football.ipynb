```python
<?php
// Check if the form was submitted with both API key and Image URL
if (isset($_GET['apiKey']) && isset($_GET['imageUrl']) && !empty(trim($_GET['imageUrl']))) {
    $apiKey   = trim($_GET['apiKey']);
    $imageUrl = trim($_GET['imageUrl']);

    // Prepare headers if an API key is provided.
    // Adjust the header field as needed by your API.
    $headers = array();
    if (!empty($apiKey)) {
        $headers[] = "x-api-key: $apiKey"; // Change to "Authorization: Bearer $apiKey" if required.
    }

    // Initialize cURL and get the image data.
    $ch = curl_init($imageUrl);
    curl_setopt($ch, CURLOPT_RETURNTRANSFER, true);
    if (!empty($headers)) {
        curl_setopt($ch, CURLOPT_HTTPHEADER, $headers);
    }
    $imageData = curl_exec($ch);

    // Check for cURL errors.
    if (curl_errno($ch)) {
        $errorMsg = curl_error($ch);
        curl_close($ch);
        die("Error making API call: " . htmlspecialchars($errorMsg));
    }

    // Get the content type (e.g., image/png or image/jpeg)
    $mimeType = curl_getinfo($ch, CURLINFO_CONTENT_TYPE);
    curl_close($ch);

    // Send the appropriate Content-Type header and output the image binary.
    header("Content-Type: " . $mimeType);
    echo $imageData;
    exit; // Terminate to avoid sending any extra HTML.
}
?>
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>Directly Serve Image from API</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            margin: 2em;
        }
        label {
            display: block;
            margin: 0.5em 0 0.2em;
        }
        input[type="text"] {
            width: 300px;
            padding: 5px;
        }
        input[type="submit"] {
            margin-top: 1em;
            padding: 5px 10px;
            font-size: 16px;
        }
    </style>
</head>
<body>
    <h1>Test Directly Serve Image</h1>
    <p>Enter your API key (if required) and the Image URL to fetch and directly serve the image.</p>
    <form method="get" action="">
        <label for="apiKey">API Key (if required):</label>
        <input type="text" name="apiKey" id="apiKey" placeholder="Enter API key">
        <br>
        <label for="imageUrl">Image URL:</label>
        <input type="text" name="imageUrl" id="imageUrl" placeholder="https://example.com/api/getImage" required>
        <br>
        <input type="submit" value="Fetch and Display Image">
    </form>
</body>
</html>
```
