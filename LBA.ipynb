## Re: Unlawful Retention of Personal Property – Claim under Tort of Conversion and Data Protection Law

Dear Sir or Madam,  

This letter constitutes formal notice of my intent to initiate legal proceedings unless an appropriate remedy is provided within 14 days. The basis of this claim concerns the prolonged, unjustified retention of my personal property between **30 May 2025** and **17 June 2025**, following the conclusion of criminal proceedings. The conduct detailed below engages the **Tort of Conversion**, breaches of **Data Protection Act 2018 / GDPR**, and non-compliance with **PACE Code B**.

---

## Chronology of Events  

- On **28 March 2024**, officers seized my personal property at the time of arrest.  
- On **30 May 2025**, I was formally notified that all charges had been dropped via a **No Further Action (NFA)** notice issued by the CPS (RF-48).  
- On **2 June 2025**, I submitted a written return request (RF-103, failed delivery).  
- On **5 June**, I issued a direct follow-up (RF-104).  
- On **10 June**, I escalated concerns to PC Heather and multiple oversight bodies (recorded in RF-104). Her email response acknowledged the Met retained possession, that no return timeline was available, and that she had no control over the evidence.  
- On **17 June 2025**, more than two weeks after the NFA, I received my property—**minus one item** (RF-108)—and witnessed a police officer tear open a sealed package despite my written objection, submitted hours earlier (RF-50). This was filmed (RF-49).  
- To date, **no seal log**, **property list**, or **review audit** has been disclosed (RF-53, RF-109).

---

## Legal Grounds

**1. Tort of Conversion**  
Unlawful possession of property post-demand and post-NFA constitutes conversion (*Kuwait Airways Corp v Iraqi Airways* [1995]). My items were lawfully seized under PACE but unlawfully retained thereafter.  

**2. Data Protection Act 2018 / GDPR**  
My GDPR **Article 21** objection was submitted on 17 June (RF-50) and ignored. Devices containing sensitive personal information (including NHS abuse disclosure, evidence bundles, and professional data) were retained and accessed without lawful basis.  

**3. PACE Code B Violations**  
No documentation of seals, itemised return logs, or custody tracking was ever provided. The breach of the sealed item on 17 June constitutes unlawful interference with evidential integrity.
