## 1. Preface: Not Past. Not Present. Still Harm

*If it still causes harm, where did that harm come from?*

Coercive control existed long before it was named in statute. What happened to me, to my children, and now to my grandchildren—wasn’t suddenly made wrong in 2015. It always was. The only thing that changed was the willingness of law to acknowledge it. And yet, the current attitude—refusing to correct falsehoods that continue to divide generations—is itself psychological abuse. There is no crime in telling the truth. But there is deep moral failure in knowing the truth, and still choosing silence. Especially when that silence punishes everyone except the person who caused the harm.

## 3. Legal Scaffold

### Key Legal Protections Violated

- **European Convention on Human Rights (ECHR):**
  - **Article 2:** Right to life (duty to act on death threats)
  - **Article 3:** Freedom from torture and inhuman treatment
  - **Article 8:** Right to family life
  - **Article 10:** Freedom of expression and truth-telling
  - **Article 13:** Right to an effective remedy

- **Domestic Legislation:**
  - **Offences Against the Person Act 1861 – Section 16:** Threats to kill
  - **Police Act 1996 – Section 39:** Duty to protect life & prevent crime
  - **Serious Crime Act 2015 – Section 76:** Controlling or coercive behaviour
  - **Serious Crime Act 2007 – Part 2:** Encouraging or assisting assault
  - **Human Rights Act 1998:** Enforces ECHR rights domestically

- **International Conventions & Oversight:**
  - **Istanbul Convention – Articles 4 & 5:** State duty to prevent and investigate
  - **UNCRC – Article 9:** Children’s right to family contact
  - **UNCAT – Article 1:** Psychological coercion as torture
  - **EU Anti-Coercion Regulation – Regulation 2023/2675**

- **Guidance & Codes:**
  - **College of Policing Code of Ethics:** Duty to safeguard the vulnerable
  - **Victims’ Code 2020 – Right 4:** Ensuring access to timely investigations and specialist support
