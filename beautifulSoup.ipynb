# BeautifulSoup

```python
from bs4 import BeautifulSoup

# Open and parse the HTML file
with open("../bookmarks.html", "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")

# Find all <a> tags and remove their 'icon' attributes
links = []
for link in soup.find_all("a"):
    if "icon" in link.attrs:
        del link["icon"]
    links.append(link["href"])  # Collect the links

# Print the links
for href in links:
    print(href)

# Save the links to a file
with open("../links.txt", "w", encoding="utf-8") as file:
    file.write("\n".join(links))

print("Icon attributes removed and links saved successfully!")
```

Steps to Build a Standalone Setup

1 Create a New Server

Set up a new Node.js project (npm init -y).

Install necessary dependencies (npm install express mongoose cors body-parser dotenv).

Organize the structure for routes, models, and config files.

2 Set Up a New Database

Use MongoDB Compass or MongoDB Atlas to create a dedicated database.

Keep this isolated from your existing collections.

Define a new collection (e.g., LinksDB).

3 Define the Schema for Your Collection

Create a model specific to categorized links.

Example schema in models/Link.js:

```python
const mongoose = require("mongoose");

const LinkSchema = new mongoose.Schema({
    url: { type: String, required: true },
    description: { type: String },
    category: { type: String, enum: ["Programming", "AWS", "Research"], required: true }
});

module.exports = mongoose.model("Link", LinkSchema);
```

4 Set Up Express Routes for Handling Requests

Add routes to insert, update, delete, and get links.

Define endpoints in routes/links.js:

```python
const express = require("express");
const Link = require("../models/Link");
const router = express.Router();

router.get("/", async (req, res) => {
    const links = await Link.find();
    res.json(links);
});

router.post("/", async (req, res) => {
    const newLink = new Link(req.body);
    await newLink.save();
    res.json(newLink);
});

router.delete("/:id", async (req, res) => {
    await Link.findByIdAndDelete(req.params.id);
    res.json({ message: "Deleted Successfully" });
});

module.exports = router;
```

5 Connect Your Standalone API to MongoDB

Define a new connection file (config/db.js) for this collection.

const mongoose = require("mongoose");
require("dotenv").config();

const connectDB = async () => {
    await mongoose.connect(process.env.MONGO_URI, {
        useNewUrlParser: true,
        useUnifiedTopology: true
    });
    console.log("MongoDB Connected...");
};

module.exports = connectDB;


6 Run Your API

Create a simple server.js file to run your API:

```python
const express = require("express");
const connectDB = require("./config/db");
const linksRoute = require("./routes/links");

require("dotenv").config();
connectDB();

const app = express();
app.use(express.json());
app.use("/api/links", linksRoute);

app.listen(5000, () => console.log("Server running on port 5000"));
```
